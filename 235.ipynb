## Context

With a few friends, you have founded a small boutique consulting firm specializing in data science and engineering.
After a few weeks of strategy work, you have decided to focus your attention on providing creditworthy assessment models for FinTech companies, and now you are trying to acquire your first client.
You want to impress potential clients with your data skills and plan to do a simple demo application of a data management system.

With your colleagues, you have split the work into two parts - data engineering and data science.
Your colleagues will make a machine learning model to predict a customer's creditworthiness.
You, on the other hand, are responsible for the data engineering part, specifically with your colleagues you agreed that you would make a data pipeline that:

1. Downloads two CSV files from [this Kaggle repository](https://www.kaggle.com/datasets/vikasukani/loan-eligible-dataset)
2. Writes the data to the database into train and test tables.

It is essential that your solution does all the tasks purely programmatically, thus you agreed to Dockerize the Python script. You also have set up the database programmatically. 
This way, you can showcase the entire solution on the clients' machines, assuming that they have Docker installed.
In addition, the containerized Python script described should be placed (publicly) in Docker Hub.

For now, you don't need to worry about how the machine learning model will be integrated.

## Project Review

During your project review, you should present it as if talking to a potential client for whom you are showcasing your solution.  
You can assume that they will have some software and data engineering skills - they will understand technical jargon but are not expected to notice and question things that could have been done better or ask about your choices.
They are very familiar with the business domain and are looking for consultants that could help them with architecting and engineering.

During the presentation, you might be asked questions to test your understanding of the covered topics, such as:

- What is the purpose of containerization?
- What advantages do containers have over virtualization technology?
- What is the relation between Kubernetes and Docker?
- What are Kubernetes Pods? How do they differ from Docker containers?
- What is a Dockerfile? Which command is used to build a Docker image from a Dockerfile?
- What is the purpose of a "Namespace" in Kubernetes?
- What is the primary purpose of the Kubernetes "etcd" component?

IMPORTANT: during the project review, you will also be asked to solve an exercise using Python.
